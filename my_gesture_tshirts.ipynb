```python
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt

# graph matrix
n_cols = 5
n_rows = 3

# generate x
# 半径
radius = 2*np.pi
x_value = np.linspace(-radius, radius, 100)
```

```python
# Graph作画関数
def graph(x, y, ax, f):
    ax.set_xlim([-radius, radius])
    ax.set_ylim([-radius, radius])
    ax.set_title(f)
    ax.grid(linewidth=1)
    ax.plot(x, y)
```

```python
# 円の作画関数
def generate_xy_for_circle(x_value, a):
    
    # zeroのベクトルを生成
    x = np.zeros(len(x_value))
    y = np.zeros(len(x_value))
    
    for i in range(len(x_value)):
        x[i] = a * np.cos(x_value[i])
        y[i] = a * np.sin(x_value[i])

    return x, y
```

```python
# 各関数を辞書型で作成
func_dict={}

func_dict["x"] = x_value
func_dict["x^2"] = pow(x_value, 2)
func_dict["x^3"]= pow(x_value, 3)
func_dict["sin x"] = np.sin(x_value)
func_dict["cos x"] = np.cos(x_value)
func_dict["tan x"]=np.tan(x_value)
func_dict["cot x"] = 1/np.tan(x_value)
func_dict["1/x"] = 1/x_value
func_dict["-1/x"]= -1/x_value
func_dict["|x|"] = np.abs(x_value)
func_dict["-|x|"] = - np.abs(x_value)
func_dict["-x"]= -x_value
func_dict["log x"] = np.log(x_value)
func_dict["a^x"] = pow(radius, x_value)
func_dict["x^2+y^2=a^2"]= 0
```

```python
# 作画
fig, ax = plt.subplots(nrows=3, ncols=5, figsize=(15, 10))

# forループを一回で回せるようにflatten()を使用
ax = ax.flatten()

i=0
for f in func_dict:
    
    x = np.zeros(len(x_value))
    y = np.zeros(len(x_value))
    
    # x yの値の取得
    if f == "x^2+y^2=a^2":
        x, y = generate_xy_for_circle(x_value, radius)
    else:
        x, y = x_value, func_dict[f]

    graph(x, y, ax[i], f)
    i=i+1
```
